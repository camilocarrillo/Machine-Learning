```python
import numpy as np
from IPython.display import display
x = np.array([[1,5,6],[4,78,9],[9,9,300]])
#create a 2D Numpy array with a diagonal of ones
eye = np.eye(4)
```

```python
print("x:\n{}".format(x))
print("Numpy array:\n{}".format(eye))
```

```python
#converte the Numpy array to Scipy sparse matrix in CSR format
from scipy import sparse
sparse_matrix = sparse.csr_matrix(eye)
print("\nScipy sparse CSR matrix:\n{}".format(sparse_matrix))
```

```python
#create sparse representation directly
data = np.ones(4)
row_i = np.arange(4)
col_i = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_i, col_i)))
print("Coo representation:\n{}".format(eye_coo))
```
