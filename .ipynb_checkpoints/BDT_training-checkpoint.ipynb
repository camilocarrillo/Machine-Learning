```python
import sys
import scipy as sp
import sklearn
import IPython
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
features = pd.read_csv("data/bdt_training_features.csv")
target = pd.read_csv("data/bdt_training_target.csv")
plt.plot(features.Cppf_phi3, features.Cppf_theta3,'--' ,marker=".")
plt.show()


```

```python
plt.hist(target.Muon_pt, bins=50, label="Muon_pt")
plt.legend()
plt.show()
```

```python
print("Keys of Features: \n{}".format(features.keys()))
print("Keys of Target: \n{}".format(target.keys()))
```

```python
features.shape
```

```python
target.shape
```

```python
#the sample preparation
from sklearn.model_selection import train_test_split
Data = pd.DataFrame(features)
Target = target.Muon_pt
X_train, X_test, y_train, y_test = train_test_split(Data, Target, test_size=0.2,random_state=0)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
# fit a linear model
from sklearn import datasets, linear_model
lm = linear_model.LinearRegression()

model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
predictions[0:5]
```

```python
print (model.score(X_test, y_test))
print (model.score(X_train, y_train))
```

```python
#the kneighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train.astype(int))

```

```python
knn.predict(X_test)
knn.score(X_train, y_train.astype(int))

```

```python
knn.score(X_test, y_test.astype(int))
```

```python
#The training
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train.astype(int))
```

```python
#prediction
prediction_tree = tree.predict(X_test)
```

```python
tree.score(X_train, y_train.astype(int))
```

```python
tree.score(X_test, y_test.astype(int))
```

```python

plt.hist(prediction_tree, bins=50, label="Prediction Muon pt")
plt.legend()
plt.show()
```
